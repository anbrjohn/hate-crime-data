```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Code to convert FBI dataset to our format

# Source: https://crime-data-explorer.fr.cloud.gov/downloads-and-docs 
# (Under additional datasets, hate crime)

# Note: https://ucr.fbi.gov/ucr-statistics-their-proper-use
# "the FBI cautions and, in fact, strongly discourages, data users 
# against using rankings to evaluate locales or the effectiveness of
# their law enforcement agencies"

# FBI raw data
df = pd.read_csv('datasets/hate_crime.csv')

# Date range
year0 = min(df['DATA_YEAR'])
yearN = max(df['DATA_YEAR'])
years = list(range(year0, yearN + 1))
print('Date range: {} to {}'.format(year0, yearN))
year2idx = dict(zip(years, range(len(years))))

# FBI's categories of victim
# Some incidents are listed as multiple biases separated by ;
all_victim_groups = sorted(set(';'.join(df['BIAS_DESC']).replace('Anti-', '').split(';')))
print()
print("FBI's Victim Groups:\n", ', '.join((all_victim_groups)))

victims = dict()
# Initialize dataset to 0 for all groups for all years
for group in all_victim_groups:
    victims[group] = np.zeros(len(years), dtype=int)
        
for year, groups, victim_count in zip(df['DATA_YEAR'], df['BIAS_DESC'], df['VICTIM_COUNT']):
    for group in groups.split(';'):
        idx = year2idx[year]
        # For every incident, increment count for group for given year by number of victims
        if victim_count == victim_count: # Skip NaN entries
            victims[group.replace('Anti-', '')][idx] += victim_count

# Convert to pandas DataFrame to export as csv file
df_victims = pd.DataFrame()
for victim_group in all_victim_groups:
    df_victims[victim_group] = victims[victim_group]
df_victims['Year'] = years
df_victims = df_victims.set_index('Year')
df_victims.to_csv('datasets/victimization_raw_numbers.csv')
```

```python
#df_victims = pd.read_csv('datasets/victimization_raw_numbers.csv').set_index('Year')
#victims = df_victims.to_dict(orient='list') # Number of hate crime victims over time

# Annual population statics for each group
df_demographics = pd.read_csv('datasets/demographics.csv').set_index('Year')

demographics = df_demographics.to_dict(orient='list') # Population size over time
years = df_demographics.index.to_list()

# Unable to get meaningful demographic data on these groups, so not included in analysis
print("Not included in analysis because of no demographic data:\n", 
      (', '.join(set(victims.keys()) - set(demographics.keys()))))

# Per Capita (per million group members) rates
df_rates = df_victims / df_demographics * 1000000
df_rates = df_rates.dropna(axis=1)
df_rates.to_csv('datasets/victimization_rates.csv')
```

```python
# From the FBI's categories

religions = ['Jewish', 'Eastern Orthodox (Russian, Greek, Other)','Sikh','Islamic (Muslim)', 
             'Catholic', 'Protestant', 'Other Christian', 'Mormon', "Jehovah's Witness",
             'Hindu','Buddhist', 'Atheism/Agnosticism', 
             'Multiple Religions, Group','Other Religion']

races = ['Arab', 'Native Hawaiian or Other Pacific Islander', 'Other Race/Ethnicity/Ancestry', 
         'Multiple Races, Group', 'Black or African American', 'American Indian or Alaska Native',
         'Asian', 'Hispanic or Latino','White']

genders = ['Lesbian (Female)','Gay (Male)', 'Transgender', 'Bisexual', 
           'Lesbian, Gay, Bisexual, or Transgender (Mixed Group)', 'Heterosexual',
           'Female', 'Male', 'Gender Non-Conforming']

disabilities = ['Physical Disability', 'Mental Disability']

selection = ['Jewish', 'Gay (Male)', 'Transgender', 'Islamic (Muslim)', 
             'Black or African American', 'American Indian or Alaska Native',
             'Asian', 'Hispanic or Latino','White']
```

```python
def get_plot(groups, per_capita=True, log=True, next_year={}):
    '''
    groups : List of names of victim groups to plot
    per_capita : Whether to show raw numbers or per 1 million 
    log : Whether the y axis scale is linear or log
    next_year : Predictions for the next year in terms of:
                {victim_group_name : percent change compared to prev year}
    '''
    markers = ['-o', '-^', '-v', '-*'] * 8
    for i, group in enumerate(groups):
        data = victims[group].copy().astype(float)
        if per_capita:
            if group not in demographics:
                print('No demographic data on:', group)
                continue
            data /= np.array(demographics[group]) / 1000000
            
        group_label = group
        if len(group_label) > 30: 
            group_label = group[:30] + '...'

        plt.plot(data, markers[i], label=group_label)
        
        if group in next_year:
            new_number = victims[group][-1] * (1 + next_year[group])
            if per_capita:
                # (Assumes same population as previous year)
                new_number /= demographics[group][-1] / 1000000
            print('New number for {}: {}'.format(group, new_number))
            # Plot the prediction year in the same color, but with a dotted line
            plt.plot([len(data)-1, len(data)], [data[-1]]+[new_number], 
                     markers[i].replace('-', ':'), 
                     c=plt.gca().lines[-1].get_color())
             

    if len(next_year) > 0:
        plt.xticks(range(len(years)+1), years+[years[-1]+1], rotation=90)    
    else:
        plt.xticks(range(len(years)), years, rotation=90)  
        
    # Hard-coded plot size 
    plt.rcParams["figure.figsize"] = (10,6)
    
    ylab = 'Number of Cases'
    if per_capita:
        ylab += ' Per Million'
    plt.ylabel(ylab)
    plt.xlabel('Year')
    #plt.title('Hate Crimes in America by Victim Group')
    if log:
        plt.yscale('log')
   
    # Magic numbers to try to get the legend box outside of the plot itself
    plt.legend(bbox_to_anchor=(1.39, 1.02), loc='upper right', ncol=1)

    return plt


# Run once to get rcParams to kick in
plt = get_plot(selection)
plt.close()
```

```python
# Not per capita
plt = get_plot(selection, per_capita=False, log=False)
plt.title('Hate Crimes in America by Targeted Group, by Total Number')
plt.savefig('graphs/selected_groups_raw_numbers.png', bbox_inches='tight')
plt.show()
```
